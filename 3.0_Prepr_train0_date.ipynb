# Data preprocessing: Date extraction

**NOTEBOOK GOAL**: Extract attributes: Day, Month and Year from the Date attribute

**DATASET TRANSFORMATION**: `(train|test).csv` >> `preprocessed0_date_(train|test).csv`

```python
# reoder variables to work on train or test dataset
work_on = 'test'
#work_on = 'train'
```

```python
from import_man import *

from datetime import datetime
```

```python
df = pd.read_csv('./dataset/' + work_on + '.csv')
df.head()
```

```python
df['D_Day'] =  df['Date'].apply(lambda x: datetime.strptime(x, '%d/%m/%Y').day)
df['D_DayOfYear'] =  df['Date'].apply(lambda x: datetime.strptime(x, '%d/%m/%Y').timetuple().tm_yday)
```

```python
df['D_Month'] =  df['Date'].apply(lambda x: datetime.strptime(x, '%d/%m/%Y').month)
```

```python
df['D_Year'] =  df['Date'].apply(lambda x: datetime.strptime(x, '%d/%m/%Y').year)
```

```python
df['D_DayOfweek'] =  df['Date'].apply(lambda x: datetime.strptime(x, '%d/%m/%Y').weekday())
```
