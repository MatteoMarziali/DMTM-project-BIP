# Data exploration - Correlation

Analysis of the attributes correlation.

```python
from dataset_features import numerical, categorical

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white", color_codes=True)
sns.set_context(rc={"font.family":'sans',"font.size":20,"axes.titlesize":30,"axes.labelsize":30})  

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('./dataset/train.csv')

# Use only numerical attributes
df = df[numerical]
df.head()
```

```python
covariance=df.corr(method='pearson')
```

```python
plt.rcParams['figure.figsize'] = [50.0,50.0]
plt.rcParams['figure.dpi']=70

sns.set(font_scale=4)
sns.heatmap(covariance,square=True, cmap="Oranges")
```

```python
plt.rcParams['figure.figsize'] = [50.0,50.0]
plt.rcParams['figure.dpi']=80

sns.set(font_scale=2)
sns.heatmap(covariance, square=True, annot=True, cmap="Oranges", xticklabels=False,  yticklabels=False)
```
