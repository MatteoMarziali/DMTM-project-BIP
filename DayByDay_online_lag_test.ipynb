```python
# calculationg R2

# calculate RSS
df_RSS = df_test
df_RSS['sqrd_diff'] = df_RSS.apply(lambda x:  (df_RSS['NumberOfSales'] - df_RSS['_NumberOfSales'])**2)
print(df_RSS.shape)
RSS = df_RSS['sqrd_diff'].sum()
print('RSS:', RSS)
df_RSS.head()
```

```python
# calculate TSS

df_TSS = df_test
pred_mean = df_TSS['_NumberOfSales'].mean()
print('pred_mean', pred_mean)

df_TSS['sqrd_diff'] = df_TSS.apply(lambda x:  (df_TSS['_NumberOfSales'] - pred_mean)**2)
TSS = df_TSS['sqrd_diff'].sum()
print('TSS:', TSS)
df_TSS.head()
```

```python
# calculate R2
print("RSS:", RSS, "   TSS:", TSS)

R2 = 1 - (RSS / TSS)
print('R_2:', R2)
```
